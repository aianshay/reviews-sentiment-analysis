```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
reviews = pd.read_csv('../data/reviews_final.csv')
reviews.head(20)
```

```python
X = reviews['review']
y = reviews['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3,
                                                    random_state=42)
```

```python
train_data = pd.concat([X_train, y_train], axis=1)

train_dataset = pd.DataFrame(train_data, columns=['review', 'label'])
train_dataset.head()
```

```python
test_data = pd.concat([X_test, y_test], axis=1)

test_dataset = pd.DataFrame(test_data, columns=['review', 'label'])
test_dataset.head(30)
```

```python
os.mkdir('../data/train')
os.chdir('../data/train')
```

```python
labels = ['Bad', 'Neutral', 'Good']

for label in labels:
  os.mkdir(label)
```

```python
count = 0

for example in train_dataset.itertuples():
  f = open(f"{example.label}/{count}.txt", "x")
  f.write(str(example.review))
  f.close()
  count += 1
```

```python
os.mkdir('../test')
os.chdir('../test')
```

```python
labels = ['Bad', 'Neutral', 'Good']

for label in labels:
  os.mkdir(label)
```

```python
count = 0

for example in test_dataset.itertuples():
  f = open(f"{example.label}/{count}.txt", "x")
  f.write(str(example.review))
  f.close()
  count += 1
```
