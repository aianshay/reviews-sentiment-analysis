```python
months = {'janeiro': '01', 'fevereiro': '02', 'mar√ßo': '03', 'abril': '04', 'maio': '05', 'junho': '06', 'julho': '07', 
         'agosto': '08', 'setembro': '09', 'outubro': '10', 'novembro': '11', 'dezembro': '12'}
```

```python
def prepro(x):
  x['title'] = x['title'].replace('\n', '')
  x['review'] = str(x['review']).replace('\n  ', '')
  x['review'] = str(x['review']).replace('\n', '')
  
  x['rating'] = x['rating'].replace(' de 5 estrelas', '')
  x['rating'] = x['rating'].replace(',', '.')

  r = re.search('de \w* de', x['date']).span()
  x['date'] = x['date'].replace(x['date'][r[0]-1:r[1]+1], '/' + months[x['date'][r[0]+3 : r[1]-3]] + '/')
  x['date'] = x['date'].replace('Avaliado no Brasil em ', '')

  return x
```

```python
df.apply(prepro, axis=1)
```

## Label Creation 

```python
reviews = pd.read_csv('../data/reviews_final.csv')
reviews.head()
```

```python
reviews.rating.unique()
```

```python
def label_creator(rating):
    if rating <= 2:
        return "Bad"
    elif rating == 3:
        return "Neutral"
    else:
        return "Good"
```

```python
reviews['label'] = reviews.rating.apply(label_creator)

reviews.head(30)
```

```python
reviews.to_csv('../data/reviews_final.csv', index=False)
```
